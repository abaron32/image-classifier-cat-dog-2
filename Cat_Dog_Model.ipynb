```python
# Step 0. Load libraries and custom modules
# System --------------------------------------------------------
import os
import pathlib
# Dataframes and matrices ---------------------------------------
import numpy as np
import pandas as pd
# Graphics ------------------------------------------------------
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
# Machine learning ----------------------------------------------
from sklearn.model_selection import train_test_split
# Deep learning -------------------------------------------------
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential # modelo secuencial que toma paso por paso para generar la arquitectura
from tensorflow.keras.layers import Conv2D,Dense,Dropout, Flatten # Convoluci칩n, dropout intencionalmente anula nodos de la red
# neuronal para que la red neuronal aprenda a superar esos problemas cuando no encuentra algo en la arquitectura
# Flatten estira la matriz en un hilo

from tensorflow.keras.layers import Activation, BatchNormalization
from tensorflow.keras.layers import MaxPooling2D # funci칩n de activacion
from tensorflow.keras import datasets, layers, models # de manera secuencial vamos agregando layers
from keras.utils import load_img
from keras.utils import img_to_array
from keras.utils import get_file
from keras.utils import image_dataset_from_directory
```

Antes de empezar, mover las carpetas Cat y Dog que hicimos a una carpeta que se llame Pets y borrar los zip.

```python
# Sample a cat image
sample_cat = mpimg.imread('Cat/image3814.jpg') #vemos la imagen usando matplotlib
plt.imshow(sample_cat)
plt.show()
```

```python
# Sample a dog image
sample_dog = mpimg.imread('Dog/image8941.jpg')
plt.imshow(sample_dog)
plt.show()
```

```python
# Count total images
data_dir = pathlib.Path('./Pets') # 
image_count = len(list(data_dir.glob('*/*.jpg')))
image_count
```

```python
# Let's follow this tutorial: https://www.tensorflow.org/tutorials/load_data/images
# Create train dataset

# Las constantes por convencion las ponemos en mayuscula
IMAGE_WITDH = 200
IMAGE_HEIGHT = 200
BATCH_SIZE = 32


# Dividimos dataset en aprendizaje y validaci칩n
# En general se divide en 3 en tareas de deeplearning: train, validation y test
# Como las 25 mil imagenes que tenemos son pocas, las de test las vamos a hacer con imagenes descargadas de google


train_ds = tf.keras.utils.image_dataset_from_directory(
  data_dir,
  validation_split=0.2,
  subset="training",
  seed=123,
  image_size=(IMAGE_WITDH, IMAGE_HEIGHT),
  batch_size=BATCH_SIZE)


  # image_dataset_from_directory: todas las imagenes que est치n dentro de una carpeta, las etiqueta con el nombre de la carpeta
```

```python
# Create validation dataset
val_ds = tf.keras.utils.image_dataset_from_directory(
  data_dir,
  validation_split=0.2, #20% de validacion
  subset="validation",
  seed=123,
  image_size=(IMAGE_WITDH, IMAGE_HEIGHT),
  batch_size=BATCH_SIZE)


# los de validacion se usan para tunear el modelo
```

```python
# Read the classes
class_names = train_ds.class_names
print(class_names)
```

```python
# Create the deep learning architecture

# las imagenes a color tienen 3 canales: RGB
IMAGE_CHANNELS=3 # tengo que decirle a la arquitectura que le mando una imagen en tres canales
IMAGE_WIDTH=200
IMAGE_HEIGHT=200


# hay arquitecturas que ya se han probado y que se ha visto que dan resultados, por ejemplo, ResNet
# https://medium.com/analytics-vidhya/understanding-resnet-architecture-869915cc2a98

# Otra es: SqueezeNet https://towardsdatascience.com/review-squeezenet-image-classification-e7414825581a
# Busca reducir la arquitectura al minimo

# Otra arquitectura: AlexNet https://www.analyticsvidhya.com/blog/2021/03/introduction-to-the-architecture-of-alexnet/

# Las arquitecturas no tienen un fundamento matem치tico directo, se han generado por ensayo y error

model = Sequential([

# 1era capa de la arquitectura, convoluci칩n que pasas por capa de activaci칩n relu
# los datos de entrada son 200*200*3 canales
# luego de la convoluci칩n hacemos una normalizaci칩n

# 32 es el kernel
Conv2D(32, (3, 3), activation='relu', input_shape=(IMAGE_WIDTH, IMAGE_HEIGHT, IMAGE_CHANNELS)),
BatchNormalization(),
MaxPooling2D(pool_size=(2, 2)),
Dropout(0.25), # es para que la red neuronal no se aprenda de memoria la estructura de los datos

Conv2D(64, (3, 3), activation='relu'),
BatchNormalization(),
MaxPooling2D(pool_size=(2, 2)),
Dropout(0.25),

Conv2D(128, (3, 3), activation='relu'),
BatchNormalization(),
MaxPooling2D(pool_size=(2, 2)),
Dropout(0.25),

Flatten(), # en esta capa tomamos la matriz cuadrada y la estiramos
Dense(512, activation='relu'),
BatchNormalization(),
Dropout(0.5),
Dense(1, activation='sigmoid'), # 2 because we have cat and dog classes
])

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
# compila el modelo
# le decimos la funci칩n de p칠rdida, para una clasificaci칩n binaria la m치s 칰til es binary_crossentropy
# optimiezer adam 

model.summary()


# 34 millones de parametros es un monton para algunas maquinas
```

```python
# Fit the model
history=model.fit(train_ds, validation_data=val_ds, epochs=10)

# validation_data para que vaya tuneando la arquitectura de red neuronal
# para que haga el aprendizaje bien le pedimos 10 pasadas (epochs)
```

```python
# Guardar el Modelo
model.save('net_model_catdog.h5')
```
